Rebuilding head-pose dataset with correct threshold

```python
# Rebuild head-pose dataset with correct threshold from your distribution

import os, shutil
import numpy as np
import cv2
import tensorflow_datasets as tfds

dataset = tfds.load("aflw2k3d", split="train")

OUT = "datasets/HEADPOSE"

# Remove old incorrect folders
shutil.rmtree(OUT, ignore_errors=True)
os.makedirs(OUT+"/on_screen", exist_ok=True)
os.makedirs(OUT+"/off_screen", exist_ok=True)

THRESHOLD = 0.095   # derived from your min/mean/max stats

count = 0

for sample in tfds.as_numpy(dataset):
    img = sample["image"]
    landmarks = sample["landmarks_68_3d_xy_normalized"]

    x_coords = landmarks[:,0]
    symmetry_score = np.std(x_coords - np.mean(x_coords))

    img = cv2.resize(img, (96,96))

    if symmetry_score < THRESHOLD:
        cv2.imwrite(f"{OUT}/on_screen/{count}.jpg", img)
    else:
        cv2.imwrite(f"{OUT}/off_screen/{count}.jpg", img)

    count += 1

print("Head-pose dataset correctly rebuilt")
```

Defining Augmentation Transforms

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# ----- Transforms -----

# Training transforms (WITH augmentation)
transform_eye_train = transforms.Compose([
    transforms.Grayscale(),
    transforms.Resize((64,64)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ToTensor()
])

# Validation & Test transforms (NO augmentation)
transform_eye_eval = transforms.Compose([
    transforms.Grayscale(),
    transforms.Resize((64,64)),
    transforms.ToTensor()
])

# Head-Pose transforms
transform_head_train = transforms.Compose([
    transforms.Grayscale(),
    transforms.Resize((64,64)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.ToTensor()
])

transform_head_eval = transforms.Compose([
    transforms.Grayscale(),
    transforms.Resize((64,64)),
    transforms.ToTensor()
])

print("All transforms defined")
```

CREATING EYE DATALOADERS

```python
# Eye datasets
eye_train_dataset = datasets.ImageFolder("datasets/MRL_Eye/train", transform=transform_eye_train)
eye_val_dataset   = datasets.ImageFolder("datasets/MRL_Eye/val", transform=transform_eye_eval)
eye_test_dataset  = datasets.ImageFolder("datasets/MRL_Eye/test", transform=transform_eye_eval)

# Eye DataLoaders
eye_train_loader = DataLoader(eye_train_dataset, batch_size=64, shuffle=True)
eye_val_loader   = DataLoader(eye_val_dataset, batch_size=64, shuffle=False)
eye_test_loader  = DataLoader(eye_test_dataset, batch_size=64, shuffle=False)

print("Eye DataLoaders created")
print("Classes:", eye_train_dataset.classes)
print("Training images:", len(eye_train_dataset))
print("Validation images:", len(eye_val_dataset))
print("Test images:", len(eye_test_dataset))
```

CREATING HEAD-POSE DATALOADER

```python
from torch.utils.data import random_split

# Full head dataset
head_full_dataset = datasets.ImageFolder("datasets/HEADPOSE", transform=transform_head_train)

# Split 80% train, 20% test
train_size = int(0.8 * len(head_full_dataset))
test_size  = len(head_full_dataset) - train_size

head_train_dataset, head_test_dataset = random_split(head_full_dataset, [train_size, test_size])

# Apply evaluation transform to test set
head_test_dataset.dataset.transform = transform_head_eval

# DataLoaders
head_train_loader = DataLoader(head_train_dataset, batch_size=64, shuffle=True)
head_test_loader  = DataLoader(head_test_dataset, batch_size=64, shuffle=False)

print("Head-Pose DataLoaders created")
print("Classes:", head_full_dataset.classes)
print("Training images:", len(head_train_dataset))
print("Test images:", len(head_test_dataset))
```

DEFINING THE CNN MODEL

```python
import torch
import torch.nn as nn

# Lightweight CNN architecture
class SmallCNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(1, 16, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(16, 32, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(),
            nn.AdaptiveAvgPool2d(1)
        )

        self.classifier = nn.Linear(64, num_classes)

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)

print("CNN model defined")
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

def draw_layer(ax, text, center, box_color):
    x, y = center
    width = 3.6
    height = 1.1

    # Rounded rectangle
    box = FancyBboxPatch(
        (x - width/2, y - height/2),
        width, height,
        boxstyle="round,pad=0.02,rounding_size=0.08",
        linewidth=2,
        edgecolor="black",
        facecolor=box_color
    )
    ax.add_patch(box)

    # Centered text
    ax.text(x, y, text, ha="center", va="center", fontsize=11, weight="bold")

# ---- Build diagram ----
fig, ax = plt.subplots(figsize=(10,8))

layers = [
    ("Input Image\n64 × 64 × 1", "#AED6F1"),
    ("Conv2D (16 filters)\nReLU", "#A9DFBF"),
    ("Max Pooling\n2 × 2", "#F9E79F"),
    ("Conv2D (32 filters)\nReLU", "#A9DFBF"),
    ("Max Pooling\n2 × 2", "#F9E79F"),
    ("Conv2D (64 filters)\nReLU", "#A9DFBF"),
    ("Global Avg Pooling", "#F5B7B1"),
    ("Dense Layer\n2 Outputs", "#D7BDE2"),
    ("Prediction\n(Awake/Sleepy or\nOn/Off Screen)", "#FAD7A0")
]

# Vertical positions
y_positions = list(reversed(range(len(layers))))

# Draw layers
for i, (layer, color) in enumerate(layers):
    draw_layer(ax, layer, (5, y_positions[i]), color)

# Draw arrows
for i in range(len(layers)-1):
    ax.annotate("",
                xy=(5, y_positions[i]-0.55),
                xytext=(5, y_positions[i+1]+0.55),
                arrowprops=dict(arrowstyle="->", linewidth=2))

ax.set_xlim(1,9)
ax.set_ylim(-1,len(layers))
ax.axis("off")

plt.title("Convolutional Neural Network Architecture", fontsize=15, weight="bold")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Device
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# Initialize models again (fresh training)
eye_model = SmallCNN(2).to(device)
head_model = SmallCNN(2).to(device)

# ----- Handle class imbalance for head-pose -----
from collections import Counter
head_labels_list = [label for _, label in head_full_dataset.samples]
count = Counter(head_labels_list)

# inverse-frequency class weights
weights = torch.tensor([1.0/count[0], 1.0/count[1]]).to(device)

# Loss functions
criterion_eye  = nn.CrossEntropyLoss()
criterion_head = nn.CrossEntropyLoss(weight=weights)

# Optimizers (lower LR for better convergence)
eye_optimizer  = optim.Adam(eye_model.parameters(), lr=0.0005)
head_optimizer = optim.Adam(head_model.parameters(), lr=0.0005)

print("Models, loss functions, and optimizers initialized")
```

Defining Loss Functions and Optimizers

TRAINING EYE-STATE CNN MODEL

```python
# ---- Train Eye Model with Early Stopping ----

max_epochs = 30        # Upper limit (won't necessarily reach this)
patience = 3          # Stop if no improvement for 3 epochs

best_val_loss = float('inf')
patience_counter = 0

for epoch in range(max_epochs):
    # --- Training ---
    eye_model.train()
    train_loss = 0

    for images, labels in eye_train_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = eye_model(images)
        loss = criterion_eye(outputs, labels)

        eye_optimizer.zero_grad()
        loss.backward()
        eye_optimizer.step()

        train_loss += loss.item()

    # --- Validation ---
    eye_model.eval()
    val_loss = 0

    with torch.no_grad():
        for images, labels in eye_val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = eye_model(images)
            loss = criterion_eye(outputs, labels)
            val_loss += loss.item()

    train_loss /= len(eye_train_loader)
    val_loss   /= len(eye_val_loader)

    print(f"Epoch {epoch+1}: Train Loss={train_loss:.4f} | Val Loss={val_loss:.4f}")

    # --- Early stopping check ---
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save(eye_model.state_dict(), "best_eye_model.pth")  # Save best model
    else:
        patience_counter += 1

    if patience_counter >= patience:
        print("Early stopping triggered — Model converged")
        break
```

```python
# Reduce learning rate for fine convergence
eye_optimizer = optim.Adam(eye_model.parameters(), lr=0.0002)

print("Learning rate reduced for fine-tuning")
```

```python
# ---- Continue Fine-Tuning Eye Model ----

extra_epochs = 10

for epoch in range(extra_epochs):
    eye_model.train()
    train_loss = 0

    for images, labels in eye_train_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = eye_model(images)
        loss = criterion_eye(outputs, labels)

        eye_optimizer.zero_grad()
        loss.backward()
        eye_optimizer.step()
        train_loss += loss.item()

    train_loss /= len(eye_train_loader)

    # Validation loss
    eye_model.eval()
    val_loss = 0
    with torch.no_grad():
        for images, labels in eye_val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = eye_model(images)
            loss = criterion_eye(outputs, labels)
            val_loss += loss.item()
    val_loss /= len(eye_val_loader)

    train_losses.append(train_loss)
    val_losses.append(val_loss)

    print(f"Fine Epoch {epoch+1}/{extra_epochs} - Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")
```

TRAINING HEAD-POSE CNN MODEL

```python
# ---- Train Head-Pose CNN ----

train_losses_head = []
val_losses_head = []

max_epochs = 40  # Upper limit
patience = 4      # Early stopping patience

best_val_loss = float('inf')
patience_counter = 0

for epoch in range(max_epochs):

    # ----- Training -----
    head_model.train()
    train_loss = 0

    for images, labels in head_train_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = head_model(images)
        loss = criterion_head(outputs, labels)

        head_optimizer.zero_grad()
        loss.backward()
        head_optimizer.step()

        train_loss += loss.item()

    train_loss /= len(head_train_loader)

    # ----- Validation (using head_test_loader as validation) -----
    head_model.eval()
    val_loss = 0

    with torch.no_grad():
        for images, labels in head_test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = head_model(images)
            loss = criterion_head(outputs, labels)
            val_loss += loss.item()

    val_loss /= len(head_test_loader)

    train_losses_head.append(train_loss)
    val_losses_head.append(val_loss)

    print(f"Epoch {epoch+1}: Train Loss={train_loss:.4f} | Val Loss={val_loss:.4f}")

    # ----- Early stopping -----
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save(head_model.state_dict(), "best_head_model.pth")
    else:
        patience_counter += 1

    if patience_counter >= patience:
        print("Head-Pose model converged — Early stopping triggered")
        break
```
